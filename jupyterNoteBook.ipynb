```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("bike-sharing-dataset.csv")
df.head()
```

```python
df = df[['date_time', 'temp', 'atemp', 'hum', 'windspeed', 'weather', 'month', 'hour', 'weekday', 'holiday', 'workingday', 'users']]
df.head()
```

```python
df['weather'].unique()
df.info()
```

```python
df['weather'] = df['weather'].map({
    'rain': 0,
    'mist': 1,
    'clear': 2
})
df.head()
```

```python
df["holiday"] = df["holiday"].astype("int64")
df["workingday"] = df["workingday"].astype("int64")
df.head()
```

```python
df["year"] = df["date_time"].dt.year
df["day"] = df["date_time"].dt.day
df["dayofweek"] = df["date_time"].dt.day_of_week
df.head(10)
```

```python
df = df[
    [
        "date_time",
        "temp",
        "atemp",
        "hum",
        "windspeed",
        "weather",
        "year",
        "month",
        "day",
        "hour",
        "dayofweek",
        "weekday",
        "holiday",
        "workingday",
        "users",
    ]
]
df.head()
```

```python
data = df.drop(['date_time','weekday'], axis=1)
data.head(10)
```

```python
plt.figure(figsize=(12, 10)) 
ax = sns.heatmap(data.corr(), annot=True)
plt.show()
```

# Build Model

```python
from sklearn.model_selection import train_test_split
```

```python
trainData = data.drop('users', axis=1)
trainData
```

```python
target = data['users']
target
```

```python
x_train, x_test, y_train, y_test = train_test_split(trainData, target, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import xgboost as xgb
regressorModel = xgb.XGBRegressor(
    n_estimators=150,
    max_depth=10,
    learning_rate=0.1
)
regressorModel.fit(x_train, y_train, eval_set=[(x_train, y_train), (x_test, y_test)])
```

```python
eval_result = regressorModel.evals_result()
training_rounds = range(len(eval_result["validation_0"]["rmse"]))
```

```python
plt.scatter(
    x=training_rounds,
    y=eval_result["validation_0"]["rmse"],
    label="Training Error",
    color="red",
)
plt.scatter(
    x=training_rounds,
    y=eval_result["validation_1"]["rmse"],
    label="Validation Error",
    color="green",
)
plt.grid(True)
plt.xlabel("Iteration")
plt.ylabel("RMSE")
plt.title("Training Vs Validation Error")
plt.legend()
plt.show()
```

```python
y_pred = regressorModel.predict(x_test)
df_compare = pd.DataFrame({'Read_Data': y_test , 'Predicted': y_pred})
df_compare
```

```python
plt.figure(figsize=(12, 10))

# Biểu đồ cho y_test
plt.subplot(2, 1, 1)
plt.plot(
    range(len(df_compare["Read_Data"])),
    df_compare["Read_Data"],
    label="y_test",
    marker="o",
)
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("y_test")
plt.legend()
plt.grid(True)

# Biểu đồ cho y_predict
plt.subplot(2, 1, 2)
plt.plot(
    range(len(df_compare["Read_Data"])),
    df_compare["Predicted"],
    label="y_predict",
    marker="x",
    color = "r"
)
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("y_predict")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
df_compare["Deviated"] = abs(df_compare['Read_Data'] - df_compare['Predicted'])
```

```python
# Vẽ histogram
plt.figure(figsize=(10, 6))
sns.histplot(df_compare["Deviated"], kde=True)
plt.title("Histogram of Deviated Values")
plt.xlabel("Deviated")
plt.ylabel("Frequency")
plt.show();
```

```python
# Vẽ histogram
plt.figure(figsize=(10, 6))
plt.plot(range(len(df_compare['Deviated'])), df_compare['Deviated'])
plt.title("Biểu đồ thể hiện sự chênh lệch")
plt.show();
```

```python
y_actual = np.array(df_compare["Read_Data"])
y_predicted = np.array(df_compare["Predicted"])
```
